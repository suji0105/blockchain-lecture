# **3. 기계학습에 필요한 기초 수학**

## 일차 함수, 기울기와 y 절편

*   **함수**란 두 집합사이의 관계를 설명하는 수학 개념 
    *  **$y = f(x)$**

*   **일차 함수**는 $y$가 $x$에 대한 일차식으로 표현된 경우
    *   **$y = ax + b (a \neq 0)$**
    *   $a$는 **기울기**, $b$는 **절편**

*   $x$가 배열인 경우, $y$도 배열
    *   **$[y] = a[x] + b (a \neq 0)$**
    *   배열 $x$에 숫자 $a$를 곱하고 숫자 $b$를 더해서 배열 $y$를 계산

```
def draw_linear_graph(a=1, b=1):
    x = np.arange(-10, 10, 0.1)
    y = [(a*num + b) for num in x]
    
    plt.plot(x,y)
    plt.ylim(-10.0,10.0)    # y축 범위 지정
    plt.grid()
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    title = 'Line equation : y = ' + str(a) + 'x + ' + str(b)
    plt.title(title)
```

```
draw_linear_graph(2,0)  
```

## 이차 함수와 최솟값

*   **이차 함수**는 $y$가 $x$에 대한 이차식으로 표현된 경우
    *   **$y = ax^{2} (a \neq 0)$**, 일반화하면 $y = a(x-p)^{2} + b (a \neq 0)$
    *   $a > 0$이면 아래로 볼록한 그래프
    *   맨 아래에 위치한지 점이 최솟값 -> 머신러닝은 이 최솟값을 찾아내는 과정임

```
def draw_quadratic_graph(a=1, p=0, b=0):
    x = np.arange(-10, 10, 0.1)
    y = [(a*(num-p)*(num-p) + b) for num in x]
    
    plt.plot(x,y)
    plt.ylim(0.0,100.0)    # y축 범위 지정
    plt.grid()
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    title = 'Quadratic equation : y = ' + str(a) + '(x-' + str(p) +')^ 2 + ' + str(b)
    plt.title(title)
```

```
draw_quadratic_graph(3,0,0) 
```

## 미분, 순간 변화율과 기울기

![대체 텍스트](http://koreaittimes.godo.co.kr/shop/data/editor/1323396041.jpg)
*    $x$의 값이 $b$ 에서 $a$까지 변할 때의 $y$의 값의 평균변화율의 극한값
![대체 텍스트](http://koreaittimes.godo.co.kr/shop/data/editor/1323395959.jpg)
    이 존재하면, 이 극한값이 함수 $y = f(x)$ 의 $x = a$ 에서의  **순간변화율** 또는 **미분계수**, **접선의 기울기**이다. 
    *    기울기가 0일 때, 즉, $x$축과 평행한 직선으로 그려질 때가 바로 그래프의 최솟값 지점이 된다.
    
*   "함수 $f(x$)를 미분하라" 
    *   $f'(x)$ 또는 $\frac{d}{dx} f(x)$로 표기
    *   $x$의 변화량이 0에 가까울 만큼 작을 때 $y$의 변화량을 $x$의 변화량으로 나누어 순간 기울기를 구하라는 의미임
*   미분의 중요한 성질
    *   $f(x) = a$ 에서 $a$가 상수일 때 미분값은 0이다.
    *   $f(x) = x$ 일 때 미분값은 1이다.
    *   $f(x) = ax$ 에서 $a$가 상수일 때 미분값은 a이다.
    *   $f(x) = x^{a}$ 에서 $a$가 자연수일 때 미분값은 $ax^{(a-1)}$이다.

```
def draw_derivative(a=1, p=0, b=0, x0=1):
    x = np.arange(-10, 10, 0.1)
    y = [(a*(num-p)*(num-p) + b) for num in x]
    
    # y = a*(x-p)^2 + b
    # y = a*(x^2 - 2px + p^2) + b
    # y' = a(2x -2p) = 2ax -2ap <- 미분식  <-x0에서의 기울기
    a1 = 2 * a * x0 - 2 * a * p    # y의 미분식 -< x0에서의 미분값이 x0에서의 접선의 기울기
    y0 = a*(x0-p)*(x0-p) + b      # x0일 때 y0를 계산, 포물선과 접손 모두 (x0, y0)를 지나감
    # y= ax + b : a1, (x0,y0)
    b1 = y0 - a1 * x0                   # 접선의 방정식을 계산하려면, 절편 계산이 필요 : y = a1x + b, b = a1x - y, (x0,y0)를 지나감
    # y= a1x + b1
    y1 = [(a1*num + b1)  for num in x]  # 접선의 방정식으로 y값 계산

    plt.plot(x, y)
    plt.plot(x, y1)
    plt.ylim(0.0,100.0)    # y축 범위 지정
    plt.xlabel("x")
    plt.ylabel("y=f(x)")
    title = 'y = ' + str(a) + '(x-' + str(p) +')^ 2 + ' + str(b) +' -- (x=' + str(x0) + ') --> y = ' + str(a1) + 'x + ' + str(b1)
    plt.title(title)
```

```
draw_derivative(2,5,3,0) # 1,0,0,5  # 2,0,0,1  # 2,0,0,5
```

## 지수와 지수 함수

*   **지수**는 $a^x$의 형태를 말함
    *   $a$를 **밑**, x를 **지수**라 함
*   **지수 함수**는 변수 $x$가 지수자리에 있는 경우를 말함
    *   $y = a^x (a\neq1, a> 0)$

```
def draw_exponential_graph(a=2):
    x = np.arange(0, 10, 0.1)
    y = [(a**num) for num in x]
    
    plt.plot(x,y)
    plt.ylim(0.0,1000.0)    # y축 범위 지정
    plt.grid()
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    title = 'Exponential equation : y = ' + str(a) + '^x'
    plt.title(title)
```

```
draw_exponential_graph(5)    # 3  # 5
```

## 시그모이드(Sigmoid) 함수

*   시그모이드 함수는 지수함수에서 밑의 값이 자연 상수인 $e$함수를 말함.
    
    $f(x) = \frac{1}{1+\exp^{-x}}$
    *   $x$가 큰 값을 가지면 $f(x)$는 1에 가까워지고
    *   $x$가 작은 값을 가지면 $f(x)$는 0에 가까와진다.
    *   S자 형태로 그려지는 이 함수의 속성은 0 또는 1, 두 개의 값 중 하나를 고를 때 유용하게 사용됨.

```
import math
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def draw_sigmoid_graph():
    x = np.arange(-10, 10, 0.1)
    y = [(sigmoid(num)) for num in x]
    plt.plot(x,y)
    plt.grid()
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    title = 'Sigmoid function : y = 1 / (1 + exp^(-x))'
    plt.title(title)
```

```
draw_sigmoid_graph()
```
